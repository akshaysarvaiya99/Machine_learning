```python
import numpy as np
from collections import Counter
import nltk

GLOVE_MODEL="./glove.6B.100d.txt"
data = "./alice_in_wonderland.txt"
seqLength = 40
verbSize = 10000
```

```python
def load_glove():
    word2em = {}
    file = open(GLOVE_MODEL, mode='rt', encoding='utf8')
    for line in file:
        words = line.strip().split()
        word = words[0]
        embeds = np.array(words[1:], dtype=np.float32)
        word2em[word] = embeds
    file.close()
    return word2em 
```

```python
word2em=load_glove()
```

```python
#input_words="Eber's Deep Conditioner every two days. And I never, ever use a blowdryer without the diffuser attachment"
input_words = input("Enter text to convert into vectors :")
input_word2em = []
encoder_input = []
for w in input_words:
    emb = np.zeros(shape=100)
    if w in word2em:
        emb = word2em[w]
    encoder_input.append(emb)

input_word2em.append(encoder_input)
```

```python
print ("Input Text :"+input_words)
print("Vectors for the input text :- ")
print(input_word2em)
```

### GLoVe word embedding for text dataset

```python
target_counter = Counter()
input_texts = []
target_texts = []
preWords = []

lines = open(data, 'rt', encoding='utf8').read().split('\n')
```

```python
#reading text file
for line in lines:
    words = [w.lower() for w in nltk.word_tokenize(line)]
    if len(words) > seqLength:
        words = words[0:seqLength]

    if len(preWords) > 0:
        input_texts.append(preWords)
        target_words = words[:]
        target_words.insert(0, 'start')
        target_words.append('end')
        for w in target_words:
            target_counter[w] += 1
        target_texts.append(target_words)
    preWords = words
```

```python
#converting word into integer based on index
word2idx = dict()
for idx, word in enumerate(target_counter.most_common(verbSize)):
    word2idx[word[0]] = idx + 1
word2idx['unknown'] = 0

#mapping id to corresponding word
idx2word = dict([(idx, word) for word, idx in word2idx.items()])
```

```python
#word embedding based on glove
texts_word2em = []

for inputWords, targetWords in zip(input_texts, target_texts):
    encoderInput = []
    for Word in inputWords:
        emb = np.zeros(shape=100)
        if Word in word2em:
            emb = word2em[Word]
        encoderInput.append(emb)
    texts_word2em.append(encoderInput)
```
